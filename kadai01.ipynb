```python
#A2.42
import numpy as np
import matplotlib.pyplot as plt
C = 15.9e-9
R = 1e+4
f = np.power(10, np.arange(12)/2)
k = 0.51
sCR = 1j * 2 * np.pi * f * C * R 
H = k * sCR / (sCR**2 + 3*sCR + 1) 
A = 6
k = 0.7
H2 = A*((sCR)**2+3*sCR+1)/((sCR)**2+(3-k*A)*sCR+1)
xx = np.real(H)
yy = np.imag(H)
x2 = np.real(H2)
y2 = np.real(H2)

gain = 20 * np.log10(np.sqrt(xx**2 + yy**2))
gain2 = 20 * np.log10(np.sqrt(x2**2 + y2**2))

fig = plt.figure(figsize = (6.4, 6.4), dpi = 300)
ax66 = fig.add_subplot(211)
ax67 = fig.add_subplot(212)
plt.subplots_adjust(hspace=0.4)

phase = np.rad2deg(np.arctan(yy/xx))
ax66.scatter(f, gain)
ax66.scatter(f, gain2)
ax66.set_xscale('log')


phase2 = np.rad2deg(np.arctan(y2/x2))
ax67.set_xscale('log')
ax67.scatter(f, phase)
ax67.scatter(f, phase2)
```
